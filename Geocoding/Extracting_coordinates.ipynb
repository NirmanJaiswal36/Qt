```python
import geopandas as gpd
import pandas as pd
from geopandas.tools import geocode
```

```python
# Reading csv file as pandas dataframe

capital_cities = pd.read_csv(r"D:\CS\GitHub\Qt\Geocoding\country-list.csv", nrows=20)
```

```python
capital_cities = capital_cities[['country','capital']]
```

```python
try:
    for index,row in capital_cities.iterrows():
        information = geocode(row['capital'] , provider = 'nominatim', user_agent = 'xyz', timeout = 15)
        capital_cities.loc[index,'Longitude'] = information.geometry.loc[0].x
        capital_cities.loc[index,'Latitude'] = information.geometry.loc[0].y
except TypeError:
    print("Coordinates of ", row["Capital"], " are not available")
```

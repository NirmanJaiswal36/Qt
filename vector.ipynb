```python
import geopandas as gpd
```

```python
#Basic plot of India with district boundaries

sf = gpd.read_file("India_shapefiles/DISTRICT_BOUNDARY.shp")
sf.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
#Using Matplotlib to plot the India's state boundaries on the axes

fig,ax = plt.subplots(figsize = (10,8))
ax.axis("off")
sb = gpd.read_file("India_shapefiles/STATE_BOUNDARY.shp")
sb.plot(ax=ax)
```

```python
#Adding color maps and legends

fig,ax = plt.subplots(figsize=(10,8))
base = sf.plot(ax=ax,column = "STATE",cmap = "jet",legend="True",legend_kwds={"loc" :"center left"})
legend = ax.get_legend()
legend.set_bbox_to_anchor((1.04,0.5))
```

```python
#Major cities plotted on the above shapefile

towns = gpd.read_file("India_shapefiles/MAJOR_TOWNS.shp")
fig,ax = plt.subplots(figsize = (20,20))
sf.plot(ax=ax,column = "STATE",cmap = "jet",legend="True",legend_kwds={"loc" :"center left"})
legend = ax.get_legend()
legend.set_bbox_to_anchor((1.04,0.5))
towns.plot(ax=ax,marker = '.',markersize = 5)
```

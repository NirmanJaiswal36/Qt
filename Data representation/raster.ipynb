```python
img = "Raster_data/LULC_2005.tif"
```

```python
ds = rasterio.open(img)
```

```python
#Reading data using Rasterio

data = ds.read()
```

```python
with rasterio.open(img) as src:
    raster_data = src.read(1)  # Read the first band
    # Get the nodata value from metadata
    nodata = src.nodata
```

```python
# Mask out the nodata values
if nodata is not None:
    data = np.ma.masked_equal(data, nodata)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(raster_data)  
plt.colorbar(label='Legend', interpolation = 'none')  # Optional: add a colorbar
plt.title('Land Use/Coverage of India')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.show()
```
